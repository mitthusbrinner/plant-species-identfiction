Detect ridges following [dsp.stackexchange.com/questions/1714/best-way-of-segmenting-veins-in-leaves](https://dsp.stackexchange.com/questions/1714/best-way-of-segmenting-veins-in-leaves) (acccessed 11 April 2023).

```python
def detect_ridges(img, sigma=2, normalize=True):
    _, maximal_eig = hessian_matrix_eigvals(hessian_matrix(img, sigma=sigma, use_gaussian_derivatives=False))
    if normalize:
        # normalize i2 values to [0, 255]
        maximal_eig = (255*(maximal_eig - maximal_eig.min())/np.ptp(maximal_eig)).astype(int) 
    return maximal_eig.astype('uint8')
```

```python
sigma = 1
ridges = detect_ridges(img, sigma=sigma)
```

```python
plt.imshow(ridges, cmap="gray")
plt.title(file_name + ' ridges')
plt.show()
```

Use [Otsu's method](https://en.wikipedia.org/wiki/Otsu%27s_method) to find optimal threshold value

```python
def convert_to_binary(img, threshold=None, log=False):
    if threshold is None:
        threshold = filters.threshold_otsu(ridges)
    if log:
        print('Optimal threshold: ' + str(threshold))
    return (img < threshold)

binarized = convert_to_binary(ridges, log=True)
```

Remove border artifact

```python
def remove_boarder(img, border_thickness):
    return img[border_thickness:-border_thickness, border_thickness:-border_thickness]

border_thickness = 5
final = remove_boarder(binarized, border_thickness)
```

```python
plt.imshow(final, cmap='gray')
plt.title(file_name + ' binarized')
plt.show()
```

## Persistent Homology 🍩

https://towardsdatascience.com/persistent-homology-with-examples-1974d4b9c3d0

Convert to point cloud data 🌥️ 

```python
point_cloud = np.argwhere(final)
print(f'points: {point_cloud.shape[0]}')
```

Choose subset randomly for performance 

```python
def random_subset(points, k, log=False):
    """
    :param: k: proportion of points to keep
    """
    n = int(point_cloud.shape[0] * k)
    if log:
        print(f'keeping {n} out of {point_cloud.shape[0]} points')
    return point_cloud[np.random.choice(point_cloud.shape[0], n, replace=False)]

k = 0.15
subset = random_subset(point_cloud, k, log=True)
```

```python
y, x = zip(*subset)
plt.gca().invert_yaxis()
plt.title(f'Sample of Point Cloud Representation \n {file_name}')
plt.scatter(x, y, s=2)
plt.show()
```


### Compute Vietoris-Rips Persistence using giotto-tda

[giotto-tda: A Topological Data Analysis Toolkit for Machine Learning and Data Exploration, Tauzin et al](https://arxiv.org/abs/2004.02551), arXiv:2004.02551, 2020.

```python
from gtda.homology import VietorisRipsPersistence
```

```python
reshaped_subset = subset[None, :, :]
```

```python
VR = VietorisRipsPersistence(homology_dimensions=[0, 1])  # homology dim 0, dim 1, ..., dim n
```

```python
tic = time.time()

diagrams = VR.fit_transform(reshaped_subset)

toc = time.time()
print(f'Done. Elapsed time: {toc - tic} seconds')
```

```python
def plot_persistence_diagram(diagram, dimensions: list, file_name: str) -> None:
    for dim in dimensions:
        diagram_dim = diagram[diagram[:,-1] == dim]  # get values corresponding to dim
        birth_death = [(birth, death) for (birth, death, _) in diagram_dim]
        birth, death = zip(*birth_death)
        plt.scatter(birth, death, label=f'$H_{dim}$')
    plt.title('Persistence Diagram of ' + file_name)
    plt.legend(loc='lower right')
    plt.xlabel('Birth')
    plt.ylabel('Death')
    X = np.linspace(0, max(max(birth), 256), 2)  # y = x line
    plt.plot(X, X, 'k--')
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
plot_persistence_diagram(diagrams[0], [0, 1], file_name)
```

## Persistent Entropy All

```python
import glob
import os
import csv

import pandas as pd

from tqdm.notebook import tqdm
```

```python
file_paths = glob.glob(input_path + "/*.JPG")
file_names = [os.path.basename(file_name) for file_name in file_paths]

entropies = []
labels = []

for file_name in tqdm(file_names):
    img = np.array(Image.open(input_path + file_name).convert('L'))
    ridges = detect_ridges(img, sigma=sigma)
    binarized = convert_to_binary(ridges)
    final = remove_boarder(binarized, border_thickness)
    point_cloud = np.argwhere(final)
    subset = random_subset(point_cloud, k)
    reshaped_subset = subset[None, :, :]
    diagrams = VR.fit_transform(reshaped_subset)
    features = PE.fit_transform(diagrams)
    
    entropies.append(tuple(features[0]))
    labels.append(int(file_name[:4]))

    print(f'{file_name} {features[0]}')
```

### Save Data as CSV 📥

```python
csv_file_name = 'plant_species_entropy.csv'

data = list(zip(labels, entropies))

with open(csv_file_name, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['species', 'H0', 'H1'])
    for row in data:
        writer.writerow([row[0], *row[1]])
```
