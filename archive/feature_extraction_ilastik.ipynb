## Load Image as Point Cloud üå•Ô∏è

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
PREPROCC_DIR = 'preprocessed_plant_leaves/'
SEGMENTED_DIR = 'segmented/'

label_to_name = {
    '0001' : 'Mango',
    '0007' : 'Pongamia Pinnata',
    '0008' : 'Basil',
    '0010' : 'Lemon'
}
```

```python
def load_image(file_name):
    cropped = cv2.imread(PREPROCC_DIR + file_name)
    segmented = cv2.imread(SEGMENTED_DIR + file_name, cv2.IMREAD_GRAYSCALE)
    return cropped, segmented
```

```python
file_name = '0008_0023.jpg'
label = file_name[:4]
cropped, segmented = load_image(file_name)
```

```python
binary_image = (segmented == 1)
```

```python
def to_point_cloud(binary_image, k, log=False):
    """
    :param: k: proportion of points to keep
    """
    point_cloud = np.argwhere(binary_image)
    n = int(point_cloud.shape[0] * k)
    if log:
        print(f'Subset consisting of {n} out of {point_cloud.shape[0]} total points')
    return point_cloud[np.random.choice(point_cloud.shape[0], n, replace=False)]
```

```python
k = 0.015
point_cloud = to_point_cloud(binary_image, k, log=True)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, sharex='col', sharey='row')
fig.suptitle('Venation Extraction ' + file_name)

axs[0, 0].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Cropped')

axs[0, 1].imshow(segmented, cmap='gray')
axs[0, 1].set_title('ilastik segmentation')

axs[1, 0].imshow(binary_image, cmap='gray')
axs[1, 0].set_title('Binary')

y, x = zip(*point_cloud)
axs[1, 1].scatter(x, y, s=1)
axs[1, 1].set_aspect('equal')
axs[1, 1].set_title(f'Point cloud $k=${k}')

plt.tight_layout()
plt.show()
```

### Skeletonise

https://scikit-image.org/docs/stable/auto_examples/edges/plot_skeleton.html

```python
from skimage.morphology import medial_axis, skeletonize, thin
```

```python
skeleton = skeletonize(binary_image)
```

```python
thinned = thin(skeleton)
```

```python
skel, distance = medial_axis(binary_image, return_distance=True)
dist_on_skel = distance * skel
dist_on_skel = (dist_on_skel > 0)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(8, 8), sharex='col', sharey='row')

# axs[0].imshow(binary_image, cmap='gray')
# axs[0].set_title('Binary')

axs[0].imshow(skeleton, cmap='gray')
axs[0].set_title('skeletonize')

axs[1].imshow(thinned, cmap='gray')
axs[1].set_title('thinned')

axs[2].imshow(dist_on_skel, cmap='gray')
axs[2].set_title('medial axis')

plt.tight_layout()
plt.show()
```

```python
k_skel = 0.4
skel_point_cloud = to_point_cloud(skeleton, k_skel, log=True)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 8), sharex='col', sharey='row')

y, x = zip(*point_cloud)
axs[0].scatter(x, y, s=1)
axs[0].invert_yaxis()
axs[0].set_aspect('equal')
axs[0].set_title(f'Point cloud $k=${k} \n {len(point_cloud)} points')

y_skel, x_skel = zip(*skel_point_cloud)
axs[1].scatter(x_skel, y_skel, s=1)
axs[1].set_aspect('equal')
axs[1].set_title(r'Point cloud skeleton $k_{skel}=$' + f'{k_skel} \n {len(skel_point_cloud)} points')

plt.show()
```

## Persistent Homology üç©

```python
import time

from gtda.homology import VietorisRipsPersistence
```

```python
reshaped_point_cloud = point_cloud[None, :, :]
```

```python
HOMOLGY_DIMS = [0, 1]
VR = VietorisRipsPersistence(homology_dimensions=HOMOLGY_DIMS)
```

```python
tic = time.time()
diagrams = VR.fit_transform(reshaped_point_cloud)
toc = time.time()
print(f'Done. Elapsed time: {toc - tic} seconds')
```

```python
def plot_persistence_diagram(diagram, dimensions: list, file_name: str):
    for dim in dimensions:
        diagram_dim = diagram[diagram[:,-1] == dim]  # get values corresponding to dim
        birth_death = [(birth, death) for (birth, death, _) in diagram_dim]
        birth, death = zip(*birth_death)
        plt.scatter(birth, death, label=f'$H_{dim}$')
    plt.title(f'Persistence Diagram {file_name}')
    plt.legend(loc='lower right')
    plt.xlabel('Birth')
    plt.ylabel('Death')
    X = np.linspace(0, max(max(birth), 256), 2)  # y = x line
    plt.plot(X, X, 'k--')
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
plot_persistence_diagram(diagrams[0], HOMOLGY_DIMS, file_name)
```

### k-dependence

```python
from scipy.optimize import curve_fit

from tqdm.notebook import tqdm
```

```python
k_values = [0.07, 0.06, 0.05, 0.04, 0.025, 0.02, 0.015, 0.010, 0.005]
k_pers_entropy = []
k_time = []
for k in tqdm(k_values):
    tic = time.time()
    point_cloud = to_point_cloud(binary_image, k)
    reshaped_point_cloud = point_cloud[None, :, :]
    diagrams = VR.fit_transform(reshaped_point_cloud)
    k_pers_entropy.append((k, PE.fit_transform(diagrams)[0]))
    toc = time.time()
    k_time.append((k, toc-tic))
```

```python
k_H0 = [(k, H0) for (k, (H0, _)) in k_pers_entropy]
k_H1 = [(k, H1) for (k, (_, H1)) in k_pers_entropy]

fig, axs = plt.subplots(nrows=1, ncols=2)
fig.suptitle(f'Persistent entropy dependence on subset size $k$ \n {file_name}')

axs[0].plot(*zip(*k_H0), 'o-', label='$H_0$')
axs[0].plot(*zip(*k_H1), 'o-', label='$H_1$')
axs[0].set_xlabel('k')
axs[0].set_ylabel('Persistent entropy')
axs[0].legend()

coeff, _ = curve_fit(lambda t, a, b, c: a * np.exp(b * t) + c, *zip(*k_time))
x_fitted = np.linspace(min(k_values), max(k_values), 100)
y_fitted = coeff[0] * np.exp(coeff[1] * x_fitted) + coeff[2]

axs[1].plot(*zip(*k_time), 'ko-')
axs[1].plot(x_fitted, y_fitted, 'k--', label='Expoential', alpha=0.4)
axs[1].set_yscale('log')
axs[1].set_xlabel('k')
axs[1].set_ylabel('Computation time (s)')
axs[1].set_ylabel('Computation time (s)')
axs[1].legend()

plt.tight_layout()
plt.show()
```

## Persistent Entropy All

```python
import glob
import os
import csv

import pandas as pd
```

Get file names for all leaves and compute entorpy for each leaf

```python
file_paths = glob.glob(SEGMENTED_DIR + "/*.jpg")
file_names = [os.path.basename(file_name) for file_name in file_paths]
```

```python
def persistent_entropy(file_name, k):
    segmented = cv2.imread(SEGMENTED_DIR + file_name, cv2.IMREAD_GRAYSCALE)
    binary_image = (segmented == 1)
    point_cloud = to_point_cloud(binary_image, k)
    reshaped_point_cloud = point_cloud[None, :, :]
    diagrams = VR.fit_transform(reshaped_point_cloud)
    return PE.fit_transform(diagrams)[0]
```

```python
entropies = []
labels = []
for file_name in tqdm(file_names):
    pe = persistent_entropy(file_name, k)
    entropies.append(pe)
    labels.append(int(file_name[:4]))
    print(f'{file_name} {pe}')
```

## Save as CSV üì•

```python
csv_file_name = 'plant_species_entropy/ilastik.csv'

data = list(zip(labels, entropies))

with open(csv_file_name, mode='w+', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Species', 'H0', 'H1'])
    for row in data:
        writer.writerow([row[0], *row[1]])
```
